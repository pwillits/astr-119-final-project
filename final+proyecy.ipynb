```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
lowerlimit = 0
upperlimit = 1.75
def cos(x):
    y = np.cos(x)
    return y
```

```python
c = 0
d = 0
def rangefinder():
    xvals = np.linspace(lowerlimit, upperlimit, 10000)
    yvals = []
    for i in xvals:
        yvals.append(cos(i))
        
    ymin = min(yvals)
    ymax = max(yvals)  
    
    global c 
    c = ymin - .5
    global d
    d = ymax + .5
    
    return c, d 
rangefinder()
```

```python
n = 10000
x = np.random.uniform(lowerlimit,upperlimit,n)
y = np.random.uniform(c,d,n)

under = np.where(cos(x) > y)[0]
over = np.where(cos(x) <= y)[0]

print (x)
print (y)
print (under)
print (over)
```

```python
frac = len(under)/(len(under)+len(over))
area = (d-c)*(upperlimit-lowerlimit)
integral = area*frac
print (frac)
print (c)
print(d)
print(area)
print (integral)
```

```python

fig = plt.figure(figsize=(7,7))
plt.xlim([lowerlimit,upperlimit])
plt.ylim([c,d])
plt.plot(x[under],y[under],'.', color = "purple")
plt.plot(x[over],y[over],'.', color = "green")
plt.plot(x,cos(x),',', color = 'black')
plt.show()
```

```python
def trapezoid_core(f,x,h):
    return 0.5*h*(f(x+h) + f(x))
```

```python
def trapezoidal_integration(f,a,b,N):
    #f == function
    #a == lower limit of integration
    #b == upper limit of integration
    #N == Number of subdivisions in range of integration
    
    #create x values on which to perform trapezoidal integration
    x = np.linspace(0, 1.75 ,n)
    h = x[1]-x[0]            #define h values
    
    #define value of the integral
    Fint = 0.0
    
    #iteration counter
    y = 0
    
    #perform the integral 
    for i in range (0,len(x)-1,1):
        Fint += trapezoid_core(f,x[i],h)
        y += 1   #add an iteration
        
    print("Found in {} iterations" .format(y))   
    #return the answer
    return Fint
```
